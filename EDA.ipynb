```python
#Let's take average
avg_opening_rank = df['opening_rank'].mean(axis = 0)
avg_open_rank = round(avg_opening_rank)
print("Avergae opening rank over the years has been - ", avg_open_rank)
```

```python
#Minimum and Maximum opening rank 
min_opening_rank = df['opening_rank'].min()
max_opening_rank = df['opening_rank'].max()
print("Minimum Opening Rank - ", min_opening_rank, " Maximum Opening Rank - ", max_opening_rank)
```

```python
plt.figure(figsize=(15,8))
year_club = df.groupby(['year', 'opening_rank']).opening_rank.count().plot(kind = 'line')
plt.legend()
```
